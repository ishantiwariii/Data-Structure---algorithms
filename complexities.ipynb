## Algorithmic complexities 

These complexities describes how the performance of an algorithm changes as the size of input changes 
 `these are of two types`
  - Time complexities 
  - Space complexities

First of all we learn about time complexities 
- Measures how the runtime of an algorithm increases with the input size.

   `Expressed using Big O notation`

### Here are three method from which we find time complexities and find which method is efficient..
  - measuring `time` to execute
  - `counting` operations involved
  - Abstract notion of `order of growth `

## Measuring time to execute 

-  Pros:

    Real-world results

    Easy to implement


- Cons:

    Depends on the machine, language, and compiler

    Affected by system load and input size

    Not useful for theoretical comparison

```python
import time 

start = time.time()
for i in range (1000):
  print(i)

print(time.time()-start)
```

## Counting operation involved 

What it is:
Count the number of basic operations (e.g., comparisons, assignments, loops) the algorithm performs.

Example:
If a loop runs n times, and inside the loop 3 operations occur ‚Üí Total = 3n

- Pros:

    More precise than just timing

    Independent of machine

- Cons:

    Tedious for complex algorithms

    Still depends on which operations are counted

    Can be messy and hard to generalize

```python
# for example 

time = 0 

for i in range (10):
  time += i

print(time)

# here the counting operation values are 
# time = 0 ->1
# for loop -> 2 operation inside and the loop has value n which is 2n where n is the value iteration of loop 
```

## Abstract Notion of Order of Growth (Big O Notation)

What it is:
Focus on how the number of operations grows as input size increases, ignoring constants and lower-order terms.

Example:

3n + 5 ‚Üí O(n)

n¬≤ + n ‚Üí O(n¬≤)

- Pros:

    Clean and general

    Shows scalability clearly

    Ignores machine-specific details

- Cons:

    Doesn't show actual performance

    Assumes "large input" behavior only

### How to calculate the big O

- Step 1: Identify Input Size

      Let the size of the input be n.

- Step 2: Look for Loops or Recursion

      Each loop that runs up to n ‚Üí contributes O(n).

      Nested loops ‚Üí multiply the complexities.

      Recursive calls ‚Üí look at how input size reduces.

- Step 3: Count the Dominant Operation

      Focus on the part of the code that grows fastest as n increases.

- Step 4: Ignore Constants

      Drop constant multipliers (like 5n becomes n).

      Drop smaller terms (like n¬≤ + n becomes n¬≤).

- Step 5: Write the Big O

      Use Big O notation: O(1), O(log n), O(n), O(n¬≤), etc.
