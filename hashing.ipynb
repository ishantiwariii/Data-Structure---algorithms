## HASHING

```python
## frequency map 

# method 1 

num = [1, 34, 21, 54, 1, 21, 53, 34, 87, 81 , 87 , 87 , 87]
freq = {}

for i in range (len(num)):
  if num[i] in freq:
    freq[num[i]] +=1
  
  else:
    freq[num[i]] = 1

print(freq)
```

```python
# method 2 

num = [1, 34, 21, 54, 1, 21, 53, 34, 87, 81 , 87 , 87 , 87]
freq = {}

for i in range (len(num)):
  freq[num[i]] = freq.get(num[i], 0) + 1

print(freq)
```

```python
# method 2 

num = [1, 34, 21, 54, 1, 21, 53, 34, 87, 81 , 87 , 87 , 87]
freq = {}

for i in num:
  freq[i] = freq.get(i, 0) + 1

print(freq)
```

### what is hashing 

- prestoring a value in any data structure and then use/fetch it .

```python
n = [1, 2, 3, 4, 5, 6, 7, 8, 1, 3, 4, 6, 6, 2]
m = [1, 25, 64, 122, 45, 6, 7, 2]

#  there are certain conditons like n contain 1 < x >= 10
# both can store 10^8 values 

hash = [0]*11

for i in n:
  hash[i] += 1

for i in m:
  if i < 1 or i > 10 :
    print("not in n")
  else :
    print(hash[i])  
```

```python
## doing that again using dict 

num = [1, 2, 3, 4, 5, 6, 7, 8, 1, 3, 4, 6, 6, 2]
mum = [1, 25, 64, 122, 45, 6, 7, 2]

#  there are certain conditons like n contain 1 < x >= 10
# both can store 10^8 values 

hash_dict = {}

for i in num:
    hash_dict[i] = hash_dict.get(i, 0) + 1

for i in mum:
    if i < 1 or i > 10:
        print(f"{i} -> not in allowed range")
    elif i in hash_dict:
        print(f"{i} -> frequency = {hash_dict[i]}")
```

```python
# finding frequency in a string

a = "abchdhweiffffhas"
b = ['a', 'b', 'c', 'f']

hash_li = [0]*26
for ch in a :
  asci = ord(ch)
  index = asci - 97

  hash_li[index] += 1

for ch in b :
  asci = ord(ch)
  index = asci - 97

  print(f"number of iteration for {ch} is {hash_li[index]}" )
```
