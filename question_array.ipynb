```python
# reversing an array

num = [2,3,45,66,6,6,2,4,7]

left = 0
right = len(num) - 1 


while left < right:
    num[left], num[right] = num[right], num[left]
    left += 1
    right -= 1

print(num)
```

```python
# defining it in a function 

def rev (num , l , r):
  while l < r:
    num[l], num[r] = num[r] , num[l]
    l = l+1 
    r = r-1
  
  return num

print(rev([23, 435, 64,2,35, 46,], 2, 5))
```

```python
def rever(num, lefti , righti):
  if lefti >= righti:
    return num 
  
  num[lefti], num[righti] = num[righti] , num[lefti]
  return rever(num, lefti+1, righti-1)

print(rever([23, 435, 64,2,35, 46,], 2, 5))
```

## largest of the array

```python
def large(num):

  larger = num[0]

  for i in range(0,len(num)):
    larger = max(larger, num[i])
  return larger

num = [34, 23, 12, 54, 22, -1]

large(num)
```

## Second largest array 

```python
def s_larger(num):

  larger = float("-inf")
  s_larger = float("-inf")

  for i in range(0,len(num)):
    larger = max(larger, num[i])

  for i in range(0,len(num)):
    if num[i] > s_larger and num[i] != larger:
      s_larger = num[i]
  
  return s_larger

num = [34, 23, 12, 54, 22, -1]

s_larger(num)
```

```python
def largest(nums):
  first = float("-inf")
  second = float("-inf")

  for num in nums:
    if num > first:
      second = first 
      first = num
    elif num > second and num != first:
      second = first 
    
  if second == float("-inf"):
    return -1 
  return second


num = [34, 23, 12, 54, 22, -1]

largest(num)
```

### moving all zeros at the end 

You are given an array arr[] of non-negative integers. You have to move all the zeros in the array to the right end while maintaining the relative order of the non-zero elements. The operation must be performed in place, meaning you should not use extra space for another array.



```python
def push_zero(num):
  if len(num) == 1 :
    return num
  
  # first we check the index of 0 and put it in the i
  i = 0 
  while i < len(num) :
    if num[i] == 0 :
      break 
    i +=1
  
  if i == len(num):
    return num
  
  # second we check the index of non zero values and then swap it to the i index and then increase the values
  j = 0 
  while j < len(num):
    if num[j] != 0 :
      num[i], num[j] = num[j] , num[i]
      i +=1
    j = j+1

  return num

num = [0, 23 , 0 , 32, 0 ,1]

push_zero(num)
```

```python
def push(num):

  if len(num) == 1:
    return num 
  
  i = 0 
  while i < len(num):
    if num[i] == 0 :
      break 
    i +=1
  
  if i == len(num):
    return num 
  
  j = i+1
  while j < len(num):
    if num[j] != 0:
      num[i], num[j] = num[j] , num[i]
      i +=1
    j +=1

  return num


num = [0, 23 , 0 , 32, 0 ,1]

push(num)
```
