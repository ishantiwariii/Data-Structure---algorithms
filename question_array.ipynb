```python
# reversing an array

num = [2,3,45,66,6,6,2,4,7]

left = 0
right = len(num) - 1 


while left < right:
    num[left], num[right] = num[right], num[left]
    left += 1
    right -= 1

print(num)
```

```python
# defining it in a function 

def rev (num , l , r):
  while l < r:
    num[l], num[r] = num[r] , num[l]
    l = l+1 
    r = r-1
  
  return num

print(rev([23, 435, 64,2,35, 46,], 2, 5))
```

```python
def rever(num, lefti , righti):
  if lefti >= righti:
    return num 
  
  num[lefti], num[righti] = num[righti] , num[lefti]
  return rever(num, lefti+1, righti-1)

print(rever([23, 435, 64,2,35, 46,], 2, 5))
```

## largest of the array

```python
def large(num):

  larger = num[0]

  for i in range(0,len(num)):
    larger = max(larger, num[i])
  return larger

num = [34, 23, 12, 54, 22, -1]

large(num)
```

## Second largest array 

```python
def s_larger(num):

  larger = float("-inf")
  s_larger = float("-inf")

  for i in range(0,len(num)):
    larger = max(larger, num[i])

  for i in range(0,len(num)):
    if num[i] > s_larger and num[i] != larger:
      s_larger = num[i]
  
  return s_larger

num = [34, 23, 12, 54, 22, -1]

s_larger(num)
```

```python
def second_largest(num):
  num.sort()
  return print(num[-2])


num = [34, 23, 12, 54, 22, -1]

second_largest(num)
```

```python
def largest(nums):
  first = float("-inf")
  second = float("-inf")

  for num in nums:
    if num > first:
      second = first 
      first = num
    elif num > second and num != first:
      second = first 
    
  if second == float("-inf"):
    return -1 
  return second


num = [34, 23, 12, 54, 22, -1]

largest(num)
```
