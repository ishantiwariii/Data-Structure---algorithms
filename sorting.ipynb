## Selection sort
Selection Sort is a comparison-based sorting algorithm that repeatedly selects the minimum element from the unsorted portion of the array and places it at the beginning.

ðŸ§  Working Principle

- Assume first element is minimum
- Find smallest element in remaining array
- Swap with current position
- Repeat for next position

â± Time Complexity
Worst Case: O(nÂ²)
Average Case: O(nÂ²)
Best Case: O(nÂ²)

```python
## selction sort in descending order 

def Selection_des(num):
  for i in range(len(num)):
    min_index = i

    for j in range(i+1, len(num)):

      if num[j] > num[min_index]:
        min_index = j
    num[i], num[min_index] = num[min_index] ,num[i]

  return num
  
num = [8, 23, 21, 1, 6, 8, 22]

Selection_des(num=num)

```

```python
## selection sort in ascending order 

def Selection_ass(num):
  n = len(num)

  for i in range(n):
    min_val = i 
    
    for j in range(i+1, n):
      if num[j] < num[min_val]:
        min_val = j 

    num[i], num[min_val] = num[min_val] ,num[i]

  return num 


num = [8, 23, 21, 1, 6, 8, 22]

Selection_ass(num)


    
```

## Bubble sort 

Bubble Sort is a simple comparison-based sorting algorithm that repeatedly compares adjacent elements and swaps them if they are in the wrong order. After each pass, the largest element "bubbles up" to its correct position at the end of the array.

ðŸ§  Working Principle

- Compare arr[j] and arr[j+1]
- Swap if arr[j] > arr[j+1]
- After first pass â†’ largest element is fixed at last position
- Repeat for remaining unsorted part


â± Time Complexity
Worst Case: O(nÂ²)
Average Case: O(nÂ²)
Best Case (optimized): O(n)

```python
def bubble_sort(num):
  n = len(num)

  for i in range(n-2, -1, -1):  
    is_swap = False
    for j in range(0 , i+1):
      if num[j] > num[j+1]:
        num[j], num[j+1] =num[j+1], num[j]
        is_swap = True
        
    if is_swap != True:
      return num


num = [8, 23, 21, 1, 6, 8, 22]
n = bubble_sort(num)
print(n)
```

## Insertion sort
Insertion Sort is a comparison-based sorting algorithm that builds the final sorted array one element at a time by inserting each element
into its correct position in the already sorted portion of the array.

ðŸ”¹ Simple Explanation

- It works the same way you arrange playing cards in your hand:
- Pick one element.
- Compare it with elements before it.
- Shift bigger elements to the right.
- Insert it at the correct position.

ðŸ”¹ Time Complexity
Best Case: O(n) (already sorted)
Average Case: O(nÂ²)
Worst Case: O(nÂ²) (reverse sorted)

```python
def insert(num):
  for i in range(1,len(num)):
    key = num[i]
    j = i-1

    while j >= 0 and num[j] > key:
      num[j+1] = num[j]
      j = j-1
    
    num[j+1] = key
  
  return num


num = [8, 23, 21, 1, 6, 7, 22]
n = insert(num)

print(n)
```
