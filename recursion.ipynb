## Recursion

- Recursion is when a function calls itself to solve a smaller version of the same problem.
- Instead of solving the whole problem at once, it keeps breaking it down into smaller pieces until it reaches a point where it can stop.

This is where the function calls itself with a smaller/simpler input.

```python
# here is an classic example of a recursion 
# here we find the factorial 

def fact (n):
  if n == 1 or n == 0 :
    return 1 
  else:
    return n*fact(n-1)
  
print(fact(7))
```

```python
# sum of first n natural number 

def add (n):
  if n == 0 :
    return 0 
  else:
    return n + add(n-1)
  
print(add(3))
```

### Here the concept below is known as head recursion 

```python
# counting to i to n 

def count (i,n):
  if i > n :
    return 
  else :
    print(i)
    count(i+1, n)

count(4,8)
```

### Here the concept below is known as tail recursion

```python
# countinng n to i 

def count(i,n):
  if i > n :
    return 
  else :
    count(i+1, n)
    print(i)

count(1,9)
```
