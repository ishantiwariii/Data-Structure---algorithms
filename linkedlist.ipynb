```python
class Node :
  def __init__(self, value):
    self.data = value 
    self.next = None
```

```python
class Linked_List :

  def __init__(self):
    
    # empty linked list 
    self.head = None
    self.n = 0 

  def __len__(self):
    return self.n
  

# inserting at the head 
  def insert_head(self, value):

    # INSERTING THE NODES
    new_node = Node(value)

    # ASSIGN THE NEXT 
    new_node.next = self.head

    # REASSIGNING THE VALUE 
    self.head = new_node

    # increment head 
    self.n = self.n + 1 

# making print method 
  def __str__(self):

    curr = self.head
    result = ''

    while curr != None:
      result = result+  str(curr.data ) + '->'
      curr = curr.next
    
    return result[:-2]

# inserting at last of linked list 
# for stopping at tail we use curr.next != None
  def append (self, value):
    new_node = Node(value)

  # CHECK IF HEAD IS EMPTY   
    if self.head == None:
      self.head = new_node
      self.n = self.n + 1
      return
    
    curr = self.head 

    while curr.next != None:
      curr = curr.next
    
    # you are at the last node 
    curr.next = new_node

    self.n = self.n + 1 

# insert at middle of linked list 
  def insert_after (self, after , value):
    new_node = Node(value)
    curr = self.head
    
    while curr != None:
      if curr.data == after:
        break
      curr = curr.next

    if curr != None:
        new_node.next = curr.next
        curr.next = new_node
    else :
        return "item not found"
```

```python
L = Linked_List()

L.insert_head(4)
L.insert_head(5)
L.insert_head(6)
L.insert_head(8)
L.append(6)
L.insert_after(6, 20)

print(L)
len(L)
```

```python
p = Linked_List()

p.append(54)
p.append(53)
p.insert_head(50)
p.insert_after(53, 456)

len(p)
print(p)
```

```python
ll = Linked_List()
ll.insert_head(10)
ll.append(20)
ll.append(30)
ll.insert_after(20, 25)
print(ll)   # Output: 10->20->25->30
print(len(ll))  # Output: 4
```
