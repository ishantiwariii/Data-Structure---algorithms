```python
# reversing an array

num = [2,3,45,66,6,6,2,4,7]

left = 0
right = len(num) - 1 


while left < right:
    num[left], num[right] = num[right], num[left]
    left += 1
    right -= 1

print(num)
```

```python
# defining it in a function 

def rev (num , l , r):
  while l < r:
    num[l], num[r] = num[r] , num[l]
    l = l+1 
    r = r-1
  
  return num

print(rev([23, 435, 64,2,35, 46,], 2, 5))
```

```python
def rever(num, lefti , righti):
  if lefti >= righti:
    return num 
  
  num[lefti], num[righti] = num[righti] , num[lefti]
  return rever(num, lefti+1, righti-1)

print(rever([23, 435, 64,2,35, 46,], 2, 5))
```
